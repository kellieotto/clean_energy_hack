```python
import pandas as pd
import string
```

```python
board_data1 = pd.read_csv("../2020_board_data.csv")
board_data1.head()
board_data2 = pd.read_csv("../corp_boards_40_80.csv", header=None)
board_data2.columns = board_data1.columns
board_data2.head()
board_data = pd.concat([board_data1, board_data2], axis=0)
```

```python
board_data.head()
```

```python
board_data['Company_clean'] = board_data['Company'].str.lower()
board_data['Company_clean'] = board_data['Company_clean'].str.replace(" /de/", "")
board_data['Company_clean'] = board_data['Company_clean'].str.replace('[^\w\s]','')
board_data['Company_clean'] = board_data['Company_clean'].str.replace(" corporation", "")
board_data['Company_clean'] = board_data['Company_clean'].str.replace(" inc", "")
board_data['Company_clean'] = board_data['Company_clean'].str.replace(" inc.", "")
board_data['Company_clean'] = board_data['Company_clean'].str.replace(" corp", "")
board_data['Company_clean'] = board_data['Company_clean'].str.replace(" ltd", "")
board_data['Company_clean'] = board_data['Company_clean'].str.replace(" co", "")


board_data.head(10)
```

```python
board_data.to_hdf('../data/board_data.h5', 'board_data')
```
